```python
import pandas as pd
import math
```

```python
df = pd.read_csv('COVID19Death_geoRegion.csv', encoding='latin-1')
df = df[df.geoRegion != 'FL']
df = df[df.datum >= '2020-03-05']
```

```python
df_g = df.groupby('datum').agg({'entries': 'sum'})

# Calc Notes per day
df_g['%'] = df_g['entries'].apply(lambda x: round(100 / 312 *x))

# BarPos
df_g = df_g.reset_index()
df_g = df_g.reset_index()
df_g['b'] = df_g['index'].apply(lambda x: x % 4 + 1)
df_g = df_g.set_index('datum')
df_g = df_g[['b', 'entries', '%']]

df_g['entries'].plot(figsize=(25, 2.5))
```
